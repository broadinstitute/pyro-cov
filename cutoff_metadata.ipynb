```python
import pandas as pd
from datetime import datetime
from tqdm import tqdm
```

```python
filename = 'metadata.tsv.gz'
cutoff = datetime.strptime('2022-01-01', '%Y-%m-%d')

df = pd.DataFrame()
for chunk in tqdm(pd.read_csv(filename, sep='\t', compression='gzip', chunksize=100000)):
    df = pd.concat([df,
                    chunk[pd.to_datetime(chunk['Collection date'], format='mixed')>=cutoff]],
                  ignore_index=True)
```

```python
df.to_csv('metadata.tsv.gz', sep='\t', compression='gzip', index=False)
```
